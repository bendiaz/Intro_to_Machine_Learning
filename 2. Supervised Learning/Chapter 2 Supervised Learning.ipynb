# Supervised Learning
## Classification and Regression

* for example: binary classification has 2 classes, a positive and negative class which classifies the object as "yes" or "no"
* multiclass is just when youhave more than 2 from a predefined set of things

* regression is concerned more with a "degree of correctness" whereas classification is yes or no

### Generalization, Overfitting, and Underfitting

* if a model is able to make accurate predictions on unseen data, we say it is able to generalize from the training set to the test set, always look for the _simplest_ model
* if we make our model too complex then we have overfit our model, which is not able to generalize well to new data
* if model is too simple, then we have underfit our data and it will also suffer in terms of performance 

### Relation of Model Complexity to Dataset Size

* the larger the dataset and more variety the more complex you can build your model without risk of overfitting

#### Analyzing KNN Classifier

* for 2D datasets we can illustrate prediction for all possible test points in xy-plane. We color it according to class, which lets us view the _decision boundary_ which is the divide between where the algo assigns class 0 vs class 1

```python
fig, axes = plt.subplots(1,3, figsize=(10,3))

for n_neighbors, ax in list(zip([1,3,9], axes)):
    #fit method returns object self, so we can instantiate
    #and fit in 1 line
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(x,y)
    mglearn.plots.plot_2d_separator(clf, x, fill=True, eps=0.5, ax=ax, alpha=0.4)
    mglearn.discrete_scatter(x[:,0],x[:,1],y,ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0 ")
    ax.set_ylabel("feature 1 ")
axes[0].legend(loc=3)
```

* (left to right) using more neighbors actually reduces model complexity

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []
#try n_neighbors from 1 to 10 lol
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    #build the model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(x_train, y_train)
    #record training set accuracy
    training_accuracy.append(clf.score(x_train, y_train))
    # record generalization accuracy
    test_accuracy.append(clf.score(x_test,y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

* he plot shows training and test set accuracy on y-axis against the setting of the n_neighbors on the x-axis.

```python
from sklearn.neighbors import KNeighborsRegressor

x,y = mglearn.datasets.make_wave(n_samples=40)

#split the wave into training and test
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 0)
#instantiate the model and set the number of neighbors to consider to 3
reg = KNeighborsRegressor(n_neighbors=3)
#fit the model using the training data and training targets
reg.fit(x_train, y_train)
```

```python
#time for predictions
print("Test set predictions \n{}".format(reg.predict(x_test)))
```

```python
#evaluate the model using the score method which returns the R^2 score
#which is known as coefficient of determination
#measures goodness of the prediction for the model (between 0 and 1)
print("test set R^2:{:.2f}".format(reg.score(x_test,y_test)))
```

#### Analyzing KNeighborsRegressor

### Linear Models

* widely used and have been studied extensively in last few decades
* model usually follows this:
$\hat{y}=w[0]*x[0]+...w[p]*x[p]+b$
    * where $x[0] ...x[p]$ denotes features of a single data point
    * $w$ and $b$ are parameters of the model that are learned and $\hat{y}$ is prediction made
        * datasets with single feature: $\hat{y}=w[0]*x[0]+b$
            * $w[0]$ is slope and b is is y-axis offset. $w$ can also be interpreted as the weight

```python
mglearn.plots.plot_linear_regression_wave()
```

* for datasets with only 1 feature this can be restrictive but for datasets with many features this can be a powerful tool. 
* if you have more features than training data points any target $y$ can be perfectly modeled as a linear function

### Linear Regression (aka ordinary least squares)

* linear regression or OLS is the simplest and most classic linear method for regression, linreg finds the parameters w and b that minimize the _mean squared error_ between predictions and the true regression targets, y, on the training set. The MSE is the sum of squared differences between predictions and the true values, divided by the number of samples. Linear regression has no parameters, which is a benefit but is also has no way to control model complexity

```python
from sklearn.linear_model import LinearRegression
x, y = mglearn.datasets.make_wave(n_samples=60)
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42)

lr = LinearRegression().fit(x_train,y_train)
```

* the slope parameters ($w$) also called _weights_ or _coefficients_, are stored in the coef_ attribute, which he offset or _intercept_ ($b$) is stored in the intercept_ attribute:

```python
print("lr.coef_:{}".format(lr.coef_))
print("lr.intercept_:{}".format(lr.intercept_))
```

* __the trailing underscore is used by scikit-learn to seperate them from parameters that are set by the user__

```python
print("training set score: {:.2f}".format(lr.score(x_train, y_train)))
print("test set score: {:.2f}".format(lr.score(x_test,y_test)))
```

```python
#not the best, but the scores are close to we are likely underfitting
#for 1-D datasets overfitting is unlikely but for higher -D datasets
#it becomes more of an issue
```

### Ridge Regression

* another linear model for regression, so formula is the same as OLS
* the difference is that there is an additional constant attached to $w$ whihc allows the magnitudes to be as close to 0 as possible, which means each feature should have as little effect on the outcome as possible, while still predicting well. 
* This constraint is an example of _regularization_ which is explicitly restricting the model to avoid overfitting, the particular kind used by ridge regression is known as __L2 regularization__

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(x_train,y_train)
print("training set score:{:.2f}".format(ridge.score(x_train,y_train)))
print("test set score:{:.2f}".format(ridge.score(x_test,y_test)))
```

```python
#training set is lower and test set is higher than with normal linreg
```

* the ridge model makes the trade-off between simplicity of the model (near 0 coefficients) and its performance on the training set
* how much importance on simplicity vs performance is set by the __alpha__ parameter, which defaults to 1.0.
* increasing alpha forces coeffs to move more toward 0, which decreases training set performance but might help generalization.

```python
ridge10 = Ridge(alpha=10.0).fit(x_train,y_train)
print("training set score:{:.2f}".format(ridge10.score(x_train,y_train)))
print("test set score:{:.2f}".format(ridge10.score(x_test,y_test)))
```

* may be too resticted

```python
ridge01 = Ridge(alpha=0.1).fit(x_train,y_train)
print("training set score:{:.2f}".format(ridge01.score(x_train,y_train)))
print("test set score:{:.2f}".format(ridge01.score(x_test,y_test)))
```

* may be too close to basic linear regression

```python
plt.plot(ridge.coef_,'s', label="ridge alpha=1")
plt.plot(ridge10.coef_,'^', label="ridge alpha=10")
plt.plot(ridge01.coef_,'v', label="ridge alpha=0.1")

plt.plot(lr.coef_, 'o',label="LinearRegression")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.hlines(0,0,len(lr.coef_))
plt.ylim(-25,25)
plt.legend()
```

* here we see that the x-axis enumerates the entries of coef_:x=0 which shows the coeff associated with the first feature and so on. 
* the main takeaway here is that for alpha=10, the coeffs are main between -3 and 3, with alpha=1 it's larger, and with 0.1 they're larger still. the linreg without regularization are outside of the chart
    * another way to understand this is to fix alpha and vary training data that's available

### Lasso

* alternative to ridge for regularizing linear regression is Lasso, as with ridge using lasso restricts coeffs to be close to 0, but with __L1 regularization__. 
* the consequence is that some coeffs become _exactly_ zero, which means some features are just ignored by the model -> can be seen as automatic feature selection: it can make model easier to interpret and can reveal the most important features of your model
    * L2 regularization penalizes the squared L2 norm of the coeffs, or the Euclidean length of w
    * L1 regularization penalizes the L1 norm of the coeff vector, the absolute value of the coeffs

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(x_train, y_train)
print("training set score: {:.2f}".format(lasso.score(x_train,y_train)))
print("test set score: {:.2f}".format(lasso.score(x_test,y_test)))
print("number of features used:{}".format(np.sum(lasso.coef_!=0)))
```

pretty terrible! this indicated we are underfitting and we find we only used 4 out of 105 features :(

```python
#let's increase max iterations:
lasso001 = Lasso(alpha=0.01,max_iter=100000).fit(x_train,y_train)
print("training set score: {:.2f}".format(lasso001.score(x_train,y_train)))
print("test set score: {:.2f}".format(lasso001.score(x_test,y_test)))
print("number of features used:{}".format(np.sum(lasso001.coef_!=0)))
```

much, much better

```python
lasso0001 = Lasso(alpha=0.0001,max_iter=100000).fit(x_train,y_train)
print("training set score: {:.2f}".format(lasso0001.score(x_train,y_train)))
print("test set score: {:.2f}".format(lasso0001.score(x_test,y_test)))
print("number of features used:{}".format(np.sum(lasso0001.coef_!=0)))
```

now we overfit :(

```python
plt.plot(lasso.coef_,'s',label='Lasso alpha=1')
plt.plot(lasso001.coef_,'^',label='Lasso alpha=0.01')
plt.plot(lasso0001.coef_,'v',label='Lasso alpha=0.0001')

plt.plot(ridge01.coef_,'o',label='ridge alpha=0.1')
plt.legend(ncol=2, loc=(0,1.05))
plt.ylim(-25,25)
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
```

* for alpha =1, we notice that most coeffs are zero, but also many are close to zero as well
* decreasing alpha to 0.01 we have the 'upward pointing triangle' which causes most features to be exactly 0
* alpha 0.0001 we get a pretty unregularized model with most coeff non-zero and with large magnitude
* ridge model (circles) all coeff are non-zero
    * in practice ridge is usually a first choice between the two, but if you have many features and only expect a few to be important than lasso is better choice. 
    * also scikit uses ElasticNet which combines the two

### Linear Models for Classification

* let's look at binary classification $\hat{y}= w[0]*x[0]+...w[p]*x[p] +b >0$
    * similar to linreg but instead of just returning the weight we have a threshold of the predicted value at zero.
    * if smaller than 0 we return -1 and larger is +1
    * $\hat{y}$ is a linear function of features: line, plane, or hyperplane(in higher dimension)
    * for linear models of classification the _decision boundary_ is a linear function of the input. 
* there are many algorithms for learning linear models, these algos all differ in the following 2 ways:
    * the way in which they measure how well a particular  combination of coefficients and interpret fits the training data
    * if and what kind of regularization they use
        different algorithms choose different ways to measure what "fitting the training set well" means. For mathmatical reasons, it is not possible to adjust $w$ & $b$ to minimize the number of miscalculations the algos produce, as one might hope. For our purposes, and many applications, different choices for item 1 in the preceeding list (called _loss functions_) are of little significance.
        
* two of the most common are _logistic regression_ and _linear support vector machines (linear SVMs)_

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
x, y = mglearn.datasets.make_forge()
fig, axes = plt.subplots(1,2, figsize=(10,3))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(x,y)
    mglearn.plots.plot_2d_separator(clf, x, fill=False, eps=0.5,
                                   ax=ax, alpha=0.7)
    mglearn.discrete_scatter(x[:,0],x[:,1],y,ax=ax)
    ax.set_title("{}".format(clf.__class__.__name__))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend()
```

* the line represents decision boundary and the upper part is 1 and lower is 0
* by default they both use L2 regularization--in the same way ridge does for regression
* for these classifiers the trade off parameter that determines the strength of the regularization is called C, and higher values of C correspond to _less_ regularization. In other words, when you use a high value for C these try to fit the training set as best as possible, while with low values of parameter C, the models put more emphasis on finding a coeff vector $w$ that is close to zero
* using low C will cause algos to try to adjust to the "majority" of data points, while using a higher value of C stresses the importance that each individual data point be classified correctly.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(cancer.data,
                                                   cancer.target,
                                                   stratify=cancer.target,
                                                   random_state=42)
logreg = LogisticRegression().fit(x_train,y_train)
print("training set score: {:.4f}".format(logreg.score(x_train,y_train)))
print("test set score: {:.4f}".format(logreg.score(x_test,y_test)))
```

with default value of C=1, let's increase C for more flexible model

```python
logreg100 = LogisticRegression(C=100).fit(x_train,y_train)
print("training set score: {:.4f}".format(logreg100.score(x_train,y_train)))
print("test set score: {:.4f}".format(logreg100.score(x_test,y_test)))
```

```python
logreg001 = LogisticRegression(C=0.01).fit(x_train,y_train)
print("training set score: {:.4f}".format(logreg001.score(x_train,y_train)))
print("test set score: {:.4f}".format(logreg001.score(x_test,y_test)))
```

with C=100 we have better accuracy so a more complex model fits better, with C=0.01 we have more underfitting occuring

let's make one last comparision

```python
plt.plot(logreg.coef_.T,'o',label="C=1")
plt.plot(logreg100.coef_.T,'^',label="C=100")
plt.plot(logreg001.coef_.T,'v',label="C=0.001")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names,rotation=90)
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.xlabel("feature")
plt.ylabel("coefficient magnitude")
plt.legend
```

* many linear classification models are for binary classification only--don't naturally extend to multiclass cases (except logistic regression) 
* a common approach is one vs rest approach, which seperates a class from all others, to make a prediction all binary classifiers are run on a test point, the one with the highest score "wins" and this class label is returned on the prediction.
* having one binary classifier per class results in having one vector of coefficients $w$ and one intercept $b$ for each class. 
    * $w[0]*x[0]+...w[p]*x[p]+b$
